```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([0,2,3,6])
ypoints = np.array([0,100,150,300])
plt.title ("Graph")
plt.xlabel("x coordinates")
plt.ylabel("y coordinates")
plt.plot(xpoints,ypoints,color = 'red',marker = 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([80,84,93,100,104,120,130,150])
ypoints = np.array([240,123,124,126,236,341,231,156])
plt.title("Group data")
plt.xlabel("Average pulse")
plt.ylabel("calorie burnage")
plt.plot(xpoints,ypoints)
plt.grid()
plt.show()
```
